```python
import pandas as pd
import pickle
from sklearn.metrics.pairwise import cosine_similarity

# Load cleaned data
df = pd.read_csv("../data/movies_cleaned.csv")

# Load vectorizer & vectors
with open("../data/tfidf_vectorizer.pkl", "rb") as f:
    tfidf = pickle.load(f)

with open("../data/movie_vectors.pkl", "rb") as f:
    movie_vectors = pickle.load(f)

print("Data and vectors loaded successfully")
```

```python
def recommend_from_dataset(movie_name, top_n=5):
    # Check if movie exists in dataset
    if movie_name not in df["title"].values:
        return None

    # Get index of movie
    idx = df[df["title"] == movie_name].index[0]

    # Compute similarity with all movies
    scores = cosine_similarity(
        movie_vectors[idx],
        movie_vectors
    ).flatten()

    # Get top similar movie indices (excluding itself)
    similar_indices = scores.argsort()[::-1][1:top_n+1]

    return df.iloc[similar_indices][["title", "vote_average"]]
```

```python
recommend_from_dataset("Avatar")
```
