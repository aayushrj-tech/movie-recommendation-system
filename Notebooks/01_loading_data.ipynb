```python
import pandas as pd

df = pd.read_csv("../data/tmdb_10000.csv")
print("Shape:", df.shape)
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["overview"])
df.reset_index(drop=True, inplace=True)

print("Rows after cleaning:", df.shape)
```

```python
df["text"] = (
    df["title"].fillna("") + " " +
    df["overview"].fillna("")
).str.lower()

df[["title", "text"]].head()
```

```python
df_cleaned = df[
    [
        "id",
        "title",
        "overview",
        "text",
        "original_language",
        "popularity",
        "vote_average",
        "vote_count"
    ]
]

df_cleaned.head()
```

```python
df_cleaned.to_csv("../Data/movies_cleaned.csv", index=False)
print("movies_cleaned.csv saved successfully")
```
